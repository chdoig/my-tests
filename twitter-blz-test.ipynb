```python
__author__ = 'Christine'

import csv
import codecs
import gzip
import xml.etree.cElementTree as et
from os.path import dirname, join
from bokeh.plotting import *

nan = float('NaN')

data = {}
with gzip.open('/Users/Christine/data/World_Country_Boundaries.csv.gz.') as f:
    decoded = codecs.iterdecode(f, "utf-8")
    next(decoded)
    reader = csv.reader(decoded, delimiter=',', quotechar='"')
    for row in reader:
        geometry, code, name = row
        xml = et.fromstring(geometry)
        lats = []
        lons = []
        for i, poly in enumerate(xml.findall('.//outerBoundaryIs/LinearRing/coordinates')):
            if i > 0:
                lats.append(nan)
                lons.append(nan)
            coords = (c.split(',')[:2] for c in poly.text.split())
            lat, lon = list(zip(*[(float(lat), float(lon)) for lon, lat in
                coords]))
            lats.extend(lat)
            lons.extend(lon)
        data[code] = {
            'name'   : name,
            'lats'   : lats,
            'lons'   : lons,
        }

#output_file('world_map.html')
output_notebook()

world_countries = data

countries_xs = [world_countries[code]['lons'] for code in world_countries]
countries_ys = [world_countries[code]['lats'] for code in world_countries]


patches(countries_xs, countries_ys, fill_alpha=0,
        fill_color="#4099FF", tools="pan,wheel_zoom,box_zoom,reset,resize,save", line_color="#000000",line_width=1,
        name="world countries", plot_width=1100, plot_height=700, title="World tweets")
show()
```
