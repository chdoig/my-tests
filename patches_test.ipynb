```python
#Get countries boundaries data
import csv
import codecs
import gzip
import xml.etree.cElementTree as et
from os.path import dirname, join

nan = float('NaN')

data = {}
with gzip.open('World_Country_Boundaries.csv.gz.') as f:
    decoded = codecs.iterdecode(f, "utf-8")
    next(decoded)
    reader = csv.reader(decoded, delimiter=',', quotechar='"')
    for row in reader:
        geometry, code, name = row
        xml = et.fromstring(geometry)
        lats = []
        lons = []
        for i, poly in enumerate(xml.findall('.//outerBoundaryIs/LinearRing/coordinates')):
            if i > 0:
                lats.append(nan)
                lons.append(nan)
            coords = (c.split(',')[:2] for c in poly.text.split())
            lat, lon = list(zip(*[(float(lat), float(lon)) for lon, lat in
                coords]))
            lats.extend(lat)
            lons.extend(lon)
        data[code] = {
            'name'   : name,
            'lats'   : lats,
            'lons'   : lons,
        }
```

```python
from bokeh.sampledata import us_states, us_counties, unemployment
from bokeh.plotting import *
from collections import OrderedDict
from bokeh.objects import (HoverTool, ColumnDataSource, Plot, Range1d, DataRange1d, FactorRange,
    Glyph, LinearAxis, DatetimeAxis, CategoricalAxis, Grid, HoverTool, DaysTicker,
    DatetimeTickFormatter, BasicTickFormatter)
from bokeh.document import Document
from bokeh.session import Session
import pandas as pd
import numpy as np
from bokeh.widgetobjects import VBoxModelForm, HBox, VBox, BokehApplet, Select, Slider, DatePicker, DateRangeSlider
from bokeh.glyphs import Patches


document = Document()
session = Session()
session.use_doc('world_tweets')
session.load_document(document)

world_countries = data.copy()
N = len(world_countries)

countries_df = pd.DataFrame.from_dict(world_countries, orient='index')

countries = sorted(countries_df.name.unique())
countries.append("All")

country = "All"


country_source = ColumnDataSource(
    data=dict(
    countries_xs = [],
    countries_ys = [],
    name = [],
    color =[],
    )
)

def make_world_plot(source):
    xdr = Range1d(start=-200, end=200)
    ydr = Range1d(start=-100, end=100)
    title = "World tweets"
    plot = Plot(title=title, data_sources=[source], x_range=xdr, y_range=ydr, plot_width=1100, plot_height=700)
    patches = Patches(xs='countries_xs', ys='countries_ys', fill_alpha='color', fill_color="#4099FF", line_color="#000000",
                      line_width=1)
    patches_glyph = Glyph(data_source=source, xdata_range=xdr, ydata_range=ydr, glyph=patches)
    plot.renderers.append(patches_glyph)

    xformatter = BasicTickFormatter (precision=None, use_scientific=False)
    yformatter = BasicTickFormatter(precision=None, use_scientific=False)
    xaxis = LinearAxis(plot=plot, dimension=0, formatter=xformatter)
    yaxis = LinearAxis(plot=plot, dimension=1, formatter=yformatter)
    xgrid = Grid(plot=plot, dimension=0, axis=xaxis)
    ygrid = Grid(plot=plot, dimension=1, axis=yaxis)
    return plot


#def make_world_plot(source):
#    figure()
#    plot = patches('countries_xs', 'countries_ys', fill_alpha='color', fill_color="#4099FF", source=source, tools="pan,wheel_zoom,box_zoom,reset,resize,hover, save", line_color="#000000",
#        line_width=1, name="world countries", plot_width=1100, plot_height=700, title="World tweets")
#    return plot

def update_data():
    
    color = np.random.random(size=N)
    
    country_source.data=dict(
        countries_xs = [world_countries[code]['lons'] for code in world_countries],
        countries_ys = [world_countries[code]['lats'] for code in world_countries],
        name = [world_countries[code]['name'] for code in world_countries],
        color = color,
    )

    session.store_document(document)
    

def on_country_change(obj, attr, old, new):
    global country
    country = new
    update_data()
    
def layout():
    #period_slider = DateRangeSlider(title="Period:", name="period", value=(start,end), bounds=(start,end), range=(dict(days=1), None))
    country_select = Select(title="Country:", value="All", options=countries)
    #language_select = Select(title="Language:", value="All", options=languages)
    
    #period_slider.on_change('value', on_period_change)
    country_select.on_change('value', on_country_change)
    #language_select.on_change('value', on_language_change)

    world_plot = make_world_plot(country_source)
    
    controls = HBox(children=[country_select])
    #controls = HBox(children=[country_select, language_select])
    layout = VBox(children=[world_plot])

    return layout
    
layout = layout()
document.add(layout)
update_data()
```
